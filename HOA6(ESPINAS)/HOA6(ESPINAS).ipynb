```python
df2 = pd.read_excel('/content/drive/MyDrive/CPE 031_ESPINAS/HOA6(ESPINAS)/SampleData.xlsx')
print(df2)
```

```python
Item_counts = df2['Item'].value_counts()

Item_counts.plot.bar(figsize=(8, 6), color='skyblue', edgecolor='black')

plt.title('Number of Items that brought by the Representative')
plt.xlabel('Items')
plt.ylabel('Total')

plt.show()
```

```python
jones_sales = df2[df2['Rep'] == 'Jones']

jones_sales = jones_sales.sort_values('OrderDate')

plt.figure(figsize=(12, 6))
plt.plot(jones_sales['OrderDate'], jones_sales['Total'], marker='o', linestyle='-')


plt.title('Sales History for Jones')
plt.xlabel('Order Date')
plt.ylabel('Total Sales ($)')


plt.xticks(rotation=45)


plt.tight_layout()


plt.show()
```
