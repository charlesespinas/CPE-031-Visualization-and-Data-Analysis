**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
# Investor audience: show relationship between marketing and sales (scatter plot)
plt.figure(figsize=(7, 4))
sns.scatterplot(x="Marketing_Spend", y="Sales", data=df, color="purple")
plt.title("Investor View: Marketing Spend vs. Sales", fontsize=13)
plt.xlabel("Marketing Spend ($)")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
#Code Here
```

```python
# Task 2: Highlight a key insight (lowest sales month)
lowest_sales_month = df["Sales"].idxmin()
colors_task2 = ["gray" if i != lowest_sales_month else "red" for i in range(len(df))]

plt.figure(figsize=(7, 4))
bars = plt.bar(df["Month"], df["Sales"], color=colors_task2)
plt.title("Highlighting the Lowest Sales Month (March)", fontsize=13)
plt.xlabel("Month")
plt.ylabel("Sales ($)")

# Add annotation for the lowest point
lowest_month_data = df.loc[lowest_sales_month]
plt.annotate(f'Lowest Sales\n(${lowest_month_data["Sales"]})',
             xy=(lowest_sales_month, lowest_month_data["Sales"]),
             xytext=(lowest_sales_month, lowest_month_data["Sales"] + 50),
             ha='center', arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

**The sales in the graph started steady in the beginning of the year, it faced a challenge in March with a dip in sales likely because to the lower marketing spend, but showed a strong recovery and growth towards the end of the period as marketing efforts increased.**

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

```python
top_10_sales = df.nlargest(10, 'Total Sales')

plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_sales['Name'], top_10_sales['Total Sales'], color='skyblue')

highest_sales_game = top_10_sales.iloc[0]
bars[0].set_color('crimson')

plt.title('Top 10 PlayStation Games by Total Sales (Highlighting the Best Seller)', fontsize=15)
plt.xlabel('Game Title', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.annotate(f"Best Seller\n({highest_sales_game['Total Sales']:.2f}M)",
             xy=(bars[0].get_x() + bars[0].get_width() / 2, bars[0].get_height()),
             xytext=(bars[0].get_x() + bars[0].get_width() / 2, bars[0].get_height() + 3000000),
             ha='center', va='bottom', color='black',
             arrowprops=dict(facecolor='black', shrink=0.05))


plt.tight_layout()
plt.show()
```

**This bar graph shows the Top 10 PlayStation games by total sales. It highlights "Grand Theft Auto V" as the best-selling game among them.**
