**3. Discussions:**

Storyboarding helps plan how your data story will unfold just like scenes in a movie.
It helps you decide what charts, annotations, or visuals are needed to communicate your message effectively.

A typical storyboard has:

Frame 1: Title or question (What are we exploring?)

Frame 2: Key data visualization or comparison

Frame 3: Main insight or conclusion

In Python, you can simulate a storyboard using multiple subplots each representing a scene or frame in your story.

**4. Procedures:**

Storyboard Example – “Sales Growth Over Time”

```python
import matplotlib.pyplot as plt

years = [2020, 2021, 2022, 2023, 2024]
sales = [100, 150, 200, 300, 450]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].text(0.5, 0.5, "Our Sales Journey\n(2020–2024)", ha='center', va='center', fontsize=14)
axes[0].axis("off")

axes[1].plot(years, sales, marker='o', color='steelblue')
axes[1].set_title("Sales Over Time")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Sales (in units)")

axes[2].text(0.5, 0.5, "Insight:\nSales grew steadily,\nwith a big jump after 2022!",
             ha='center', va='center', fontsize=12, color='darkgreen')
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

**5. Supplementary Activity:**

1. Choose a small dataset (e.g., from Kaggle or manually created).
Then, create a 3-frame storyboard in Python showing:

Frame 1: Title or question

Frame 2: One simple chart (bar or line)

Frame 3: Main message or conclusion

2. Use matplotlib and label each part clearly.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [120, 150, 170, 130, 180, 200]
}
df = pd.DataFrame(data)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].text(0.5, 0.5,
             "Which Month Has the Highest Sales?",
             ha='center', va='center', fontsize=12, wrap=True)
axes[0].set_title("Frame 1: Title or Question")
axes[0].axis('off')

axes[1].bar(df['Month'], df['Sales'], color='skyblue')
axes[1].set_title("Frame 2: Sales per Month")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Sales")

highest_month = df.loc[df['Sales'].idxmax(), 'Month']
axes[2].text(0.5, 0.5,
             f"The highest sales occurred in {highest_month}!",
             ha='center', va='center', fontsize=12, wrap=True, color='green')
axes[2].set_title("Frame 3: Main Message")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
