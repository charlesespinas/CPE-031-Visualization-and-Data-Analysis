**3. Discussions:**

A well-designed data dashboard is more than a collection of charts — it’s a purposeful storytelling tool that connects data to decisions.

Understanding the Purpose of the Data Dashboard:
Every dashboard must have a clear objective — whether it’s to monitor performance, analyze patterns, or inform strategic decisions. The design should align with this goal, focusing only on the most relevant data and metrics.

Considering the Needs of the Data Dashboard’s Users:
Dashboards should be designed with the user in mind. A manager may need high-level KPIs, while an analyst may need granular, filterable data. Knowing the user’s goals, skills, and workflow ensures that the dashboard supports, rather than overwhelms, their decision-making.

Data Dashboard Engineering:
Engineering a dashboard involves organizing visual elements efficiently, ensuring smooth interactivity, and maintaining data accuracy. It also includes performance considerations (like handling large datasets), responsive design, and consistent color schemes.

Step 2: Create a Sample Dataset

We’ll simulate student performance data to design an academic performance dashboard.

```python
# Create dataset
np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)
df.head()
```

Step 3: Design Dashboard Components

```python
# Average Score per Subject
fig1 = px.bar(df.groupby("Subject", as_index=False)["Score"].mean(),
              x="Subject", y="Score", title="Average Score per Subject",
              color="Score", color_continuous_scale="Blues")

# Attendance vs. Score Scatter
fig2 = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                  title="Attendance vs. Performance", hover_data=["Student"])

# Score Distribution
fig3 = px.box(df, x="Subject", y="Score", title="Score Distribution by Subject",
              color="Subject")

fig1.show()
fig2.show()
fig3.show()
```

Step 4: Combine into One Interactive Dashboard

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Aggregated data
avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

# Dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score",
                                    "Score Distribution", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

fig.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)
fig.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="Scatter Plot"), row=1, col=2)
fig.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig.update_layout(height=700, width=950, title_text="🎓 Student Performance Dashboard")
fig.show()
```

**Task 2:** Design for User Needs

Modify the dashboard to meet specific user needs:

If your user is a teacher, highlight underperforming students.

If your user is a student, highlight their personal progress vs. class average.

If your user is a department head, summarize subject-level averages only.

```python
# Code Here
from plotly.subplots import make_subplots
import plotly.graph_objects as go

np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)
df.head()

# Task 2: Design for User Needs

# Teacher View: Highlight underperforming students (e.g., those with scores below 70)
underperforming_students = df[df['Score'] < 70]
if not underperforming_students.empty:
    fig_teacher = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                             title="Teacher View: Underperforming Students Highlighted",
                             hover_data=["Student"])
    fig_teacher.add_scatter(x=underperforming_students["Attendance (%)"],
                            y=underperforming_students["Score"],
                            mode="markers",
                            marker=dict(color='red', size=10, symbol='x'),
                            name='Underperforming')
    fig_teacher.show()
else:
    print("No underperforming students found based on the criteria.")


# Student View: Highlight a specific student's progress vs. class average
# Let's pick a sample student, e.g., "Student 10"
sample_student = "Student 10"
student_data = df[df["Student"] == sample_student]
class_avg_score = df["Score"].mean()

if not student_data.empty:
    fig_student = px.bar(student_data, x="Subject", y="Score",
                         title=f"Student View: {sample_student}'s Scores vs. Class Average",
                         color="Subject")
    fig_student.add_shape(type="line",
                          x0=-0.5, y0=class_avg_score, x1=len(student_data["Subject"]) - 0.5, y1=class_avg_score,
                          line=dict(color="Red", dash="dash"),
                          name="Class Average")
    fig_student.show()
else:
    print(f"No data found for {sample_student}.")

# Department Head View: Summarize subject-level averages only
subject_avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
subject_avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

fig_dept_head = make_subplots(rows=1, cols=2,
                              subplot_titles=("Average Score per Subject (Department Head)",
                                              "Average Attendance per Subject (Department Head)"),
                              specs=[[{"type": "bar"}, {"type": "bar"}]])

fig_dept_head.add_trace(go.Bar(x=subject_avg_scores["Subject"], y=subject_avg_scores["Score"], name="Average Score"), row=1, col=1)
fig_dept_head.add_trace(go.Bar(x=subject_avg_attendance["Subject"], y=subject_avg_attendance["Attendance (%)"], name="Average Attendance"), row=1, col=2)

fig_dept_head.update_layout(height=400, width=950, title_text="🏢 Department Head View: Subject Level Averages")
fig_dept_head.show()
```

**5. Supplementary Activity:**

Design Your Own Dashboard: “Smart Campus Energy Monitor”
Create a dashboard that shows the energy usage of different campus areas (e.g., laboratories, offices, classrooms) over time.

1. Simulate or use sample data (Energy consumption per hour or per day).

2. Include at least three charts (bar, line, pie).

3. Ensure your design reflects a clear purpose and target user (e.g., facilities manager).

4. Write a short explanation of your engineering decisions (why you chose certain colors, layouts, or interactivity).

```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
campus_areas = ["Laboratories", "Offices", "Classrooms", "Library", "Cafeteria"]

data = {
    "Date": np.random.choice(dates, 200),
    "Campus Area": np.random.choice(campus_areas, 200),
    "Energy Consumption (kWh)": np.random.uniform(50, 500, 200).round(2)
}

energy_df = pd.DataFrame(data)

daily_energy = energy_df.groupby("Date", as_index=False)["Energy Consumption (kWh)"].sum()
area_energy = energy_df.groupby("Campus Area", as_index=False)["Energy Consumption (kWh)"].sum()


fig1 = px.line(daily_energy, x="Date", y="Energy Consumption (kWh)",
               title="Daily Energy Consumption Over Time")

fig2 = px.bar(area_energy, x="Campus Area", y="Energy Consumption (kWh)",
               title="Total Energy Consumption per Campus Area",
               color="Campus Area", color_continuous_scale="Viridis")

fig3 = px.pie(area_energy, values="Energy Consumption (kWh)", names="Campus Area",
               title="Proportion of Energy Consumption by Campus Area")
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Daily Energy Consumption Over Time", "Total Energy Consumption per Campus Area",
                                    "Proportion of Energy Consumption by Campus Area", ""),
                    specs=[[{"type": "xy"}, {"type": "bar"}],
                           [{"type": "pie"}, {"type": "scatter"}]])

fig.add_trace(fig1['data'][0], row=1, col=1)
fig.add_trace(fig2['data'][0], row=1, col=2)
fig.add_trace(fig3['data'][0], row=2, col=1)


fig.update_layout(height=800, width=1000, title_text="💡 Smart Campus Energy Monitor")
fig.show()
```

4.Engineering Decisions Explanation:

For the "Smart Campus Energy Monitor"dashboard, the reason why we choose this layoute is to present the key information efficiently to a facilities manager. The line chart that we make shows the daily energy consumption over time is placed prominently to quickly and immediately identify the trends and anomalies. On the other hand, the bar chart and the pie chart that we make provides the breakdown of the energy usage by campus area which allows for the easy comparison and identification of areas with high consumption. We chose this color palettes for the clarity and visual appeal using a sequential scale for the bar chart to indicate the consumption levels and distinction colors for the pie chart slices to differentiate areas. WThe charts let you hover over them to see more details and in the future, we could add ways to filter or look closer at specific areas or times.
